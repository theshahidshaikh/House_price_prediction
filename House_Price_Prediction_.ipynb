```python
#Importing depandencis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
#importing dataset
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
house_price_dataframe = pd.DataFrame(house_price_dataset.data,columns = house_price_dataset.feature_names)
```

```python
print(house_price_dataframe)
```

```python
#adding price cloumn in dataset (target)
house_price_dataframe['Price'] = house_price_dataset.target
```

```python
print(house_price_dataframe)
```

```python
house_price_dataframe.isnull().sum()
```

```python
house_price_dataframe.describe()
```

```python
correlation = house_price_dataframe.corr()
print(correlation)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.2f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
X = house_price_dataframe.drop(['Price'],axis=1)
Y = house_price_dataframe['Price']
print(X)
print(Y)
```

```python
#dividing data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
#Model training
#XGBoost Regressor
model = XGBRegressor()
model.fit(X_train,Y_train)
```

```python
#model evoluation
#predection training
model_predicition_training = model.predict(X_train)
#R squared
score_1 = metrics.r2_score(Y_train,model_predicition_training)
#mean absolute error
score_2 = metrics.mean_absolute_error(Y_train,model_predicition_training)
print("R squared error : ",score_1)
print("Mean absolute error : ",score_2)
```

```python
#predection test
model_predicition_test = model.predict(X_test)
#R squared
score_t1 = metrics.r2_score(Y_test,model_predicition_test)
#mean absolute error
score_t2 = metrics.mean_absolute_error(Y_test,model_predicition_test)
print("R squared error : ",score_t1)
print("Mean absolute error : ",score_t2)
```

```python
#visiulizing
plt.scatter(Y_train,model_predicition_training)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual price vs Predicted price")
plt.show()
```
